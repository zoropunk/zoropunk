### Creating Indexes: Summary

#### 1. **TF-IDF (Term Frequency-Inverse Document Frequency):**
   - **Concept:** Measures the importance of words in a document based on their frequency and rarity across documents.
   - **Usage:** Basic search and retrieval tasks.
   - **Pros:** Simple, efficient.
   - **Cons:** Lacks semantic understanding.


#### 2. **Word2Vec (Word Embeddings):**
   - **Concept:** Converts words into continuous vectors that capture semantic relationships.
   - **Usage:** Understanding word similarities and context.
   - **Pros:** Captures semantic meaning.
   - **Cons:** Requires a large corpus; limited to known vocabulary.

#### 3. **Transformers (e.g., BERT, GPT):**
   - **Concept:** Uses self-attention to create contextualized word embeddings.
   - **Usage:** Complex tasks like question-answering and semantic search.
   - **Pros:** Rich contextual understanding.
   - **Cons:** Computationally expensive.

#### 4. **LLMs (Large Language Models):**
   - **Concept:** Pre-trained models that encode complex semantic meanings.
   - **Usage:** Advanced AI applications requiring deep language understanding.
   - **Pros:** Highly powerful and versatile.
   - **Cons:** Resource-intensive; potential biases.

### Choosing the Right Method
- **TF-IDF:** Quick, simple searches.
- **Word2Vec/Transformers:** Need for semantic/contextual understanding.
- **LLMs:** Advanced, nuanced tasks.

```python
def get_datafarme(path):

  #get total no of files in a direcory
  _ , _,  files = next(os.walk(path))

  #all files in search engine data
  filePath = path + "/*.***"

  #get directory
  filesList = sorted(glob.glob(filePath))

  # store in dataframe
  fileName = [f for f in filesList]

  #creating ids
  id = [i for i in range(len(files))]

  #get file type
  filetype = [types.split(".")[-1] for types in fileName]

  #create pandas dataframe
  df = pd.DataFrame({'id': id ,
            "fileName" : fileName,
            "fileType" : filetype})

  return df

data = get_datafarme("/content/books")

```

```python
import re
from tqdm import tqdm

# Precompile regex for decontraction
decontracted_re = [
    (re.compile(r"won't"), "will not"),
    (re.compile(r"can\'t"), "can not"),
    (re.compile(r"n\'t"), " not"),
    (re.compile(r"\'re"), " are"),
    (re.compile(r"\'s"), " is"),
    (re.compile(r"\'d"), " would"),
    (re.compile(r"\'ll"), " will"),
    (re.compile(r"\'t"), " not"),
    (re.compile(r"\'ve"), " have"),
    (re.compile(r"\'m"), " am"),
]

def decontracted(phrase):
    for pattern, repl in decontracted_re:
        phrase = pattern.sub(repl, phrase)
    return phrase

# Predefined stopwords list (customized)
stopwords = set([
    'i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', "you're",
    "you've", "you'll", "you'd", 'your', 'yours', 'yourself', 'yourselves', 'he',
    'him', 'his', 'himself', 'she', "she's", 'her', 'hers', 'herself', 'it',
    "it's", 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves',
    'what', 'which', 'who', 'whom', 'this', 'that', "that'll", 'these', 'those',
    'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had',
    'having', 'do', 'does', 'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if',
    'or', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for', 'with',
    'about', 'against', 'between', 'into', 'through', 'during', 'before', 'after',
    'above', 'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 'off', 'over',
    'under', 'again', 'further', 'then', 'once', 'here', 'there', 'when', 'where',
    'why', 'how', 'all', 'any', 'both', 'each', 'few', 'more', 'most', 'other',
    'some', 'such', 'only', 'own', 'same', 'so', 'than', 'too', 'very', 's', 't',
    'can', 'will', 'just', 'don', "don't", 'should', "should've", 'now', 'd', 'll',
    'm', 'o', 're', 've', 'y', 'ain', 'aren', "aren't", 'couldn', "couldn't", 'didn',
    "didn't", 'doesn', "doesn't", 'hadn', "hadn't", 'hasn', "hasn't", 'haven',
    "haven't", 'isn', "isn't", 'ma', 'mightn', "mightn't", 'mustn', "mustn't",
    'needn', "needn't", 'shan', "shan't", 'shouldn', "shouldn't", 'wasn', "wasn't",
    'weren', "weren't", 'won', "won't", 'wouldn', "wouldn't"
])

def preprocess_text(text_data):
    preprocessed_text = []
    for sentence in tqdm(text_data):
        if not sentence:  # Skip None or empty inputs
            continue

        sentence = decode_text(sentence)
        # Decontract
        sentence = decontracted(sentence)

        # Replace escape sequences and unwanted characters
        sentence = sentence.replace('\\r', ' ').replace('\\n', ' ').replace('\\"', ' ')
        sentence = re.sub('[^A-Za-z0-9]+', ' ', sentence)

        # Remove stopwords and lowercase
        sentence = ' '.join(word for word in sentence.split() if word.lower() not in stopwords)
        preprocessed_text.append(sentence.lower().strip())

    return preprocessed_text

def decode_text(text_bytes):
    try:
        # Decode bytes to string
        return text_bytes.decode('utf-8')
    except (UnicodeDecodeError, AttributeError):
        # Handle cases where text is not bytes
        return text_bytes
```

```python
def similar_files(query):
  """
  transform query,
  calculate cosine similarity,
  get sorted index and tf idf socre ,
  use index to call data

  """
  from sklearn.metrics.pairwise import cosine_similarity
  from sklearn.feature_extraction.text import TfidfVectorizer
  import joblib
  import re

  tfidf_pickle_vector = joblib.load('/content/drive/MyDrive/search_engine/tfidf_vectorizer.pkl')
  tfidf_pickle_score = joblib.load('/content/drive/MyDrive/search_engine/tfidf.pkl')




  stopwords= ['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', "you're", "you've",\
              "you'll", "you'd", 'your', 'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', \
              'she', "she's", 'her', 'hers', 'herself', 'it', "it's", 'its', 'itself', 'they', 'them', 'their',\
              'theirs', 'themselves', 'what', 'which', 'who', 'whom', 'this', 'that', "that'll", 'these', 'those', \
              'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'having', 'do', 'does', \
              'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until', 'while', 'of', \
              'at', 'by', 'for', 'with', 'about', 'against', 'between', 'into', 'through', 'during', 'before', 'after',\
              'above', 'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 'off', 'over', 'under', 'again', 'further',\
              'then', 'once', 'here', 'there', 'when', 'where', 'why', 'how', 'all', 'any', 'both', 'each', 'few', 'more',\
              'most', 'other', 'some', 'such', 'only', 'own', 'same', 'so', 'than', 'too', 'very', \
              's', 't', 'can', 'will', 'just', 'don', "don't", 'should', "should've", 'now', 'd', 'll', 'm', 'o', 're', \
              've', 'y', 'ain', 'aren', "aren't", 'couldn', "couldn't", 'didn', "didn't", 'doesn', "doesn't", 'hadn',\
              "hadn't", 'hasn', "hasn't", 'haven', "haven't", 'isn', "isn't", 'ma', 'mightn', "mightn't", 'mustn',\
              "mustn't", 'needn', "needn't", 'shan', "shan't", 'shouldn', "shouldn't", 'wasn', "wasn't", 'weren', "weren't", \
              'won', "won't", 'wouldn', "wouldn't"]

      # tqdm is for printing the status bar
  phrase = re.sub(r"n\'t", " not", query)
  phrase = re.sub(r"\'re", " are", phrase)
  phrase = re.sub(r"\'s", " is", phrase)
  phrase = re.sub(r"\'d", " would", phrase)
  phrase = re.sub(r"\'ll", " will", phrase)
  phrase = re.sub(r"\'t", " not", phrase)
  phrase = re.sub(r"\'ve", " have", phrase)
  phrase = re.sub(r"\'m", " am", phrase)
  sent = phrase.replace('\\r', ' ')
  sent = sent.replace('\\n', ' ')
  sent = sent.replace('\\"', ' ')
  sent = re.sub('[^A-Za-z0-9]+', ' ', sent)
  # https://gist.github.com/sebleier/554280
  sent = ' '.join(e for e in sent.split() if e.lower() not in stopwords)



  query_vec = tfidf_pickle_vector.transform([sent])
  results = cosine_similarity(tfidf_pickle_score,query_vec).reshape((-1,))
  a = list(enumerate(results))
  a.sort(key=lambda k: k[1], reverse=True)

  top_10_files = [i[0] for i in a[:10]]
  return df.fileName.iloc[top_10_files]
```
